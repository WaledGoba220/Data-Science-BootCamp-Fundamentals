## <span style='color:#2E8BC0'> 0- import Libraries </span>

```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

## <span style='color:#2E8BC0'> 1- Read Data </span>

```python
dataset  =pd.read_csv('gld_price_data.csv')
df = dataset.copy()
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

## <span style='color:#2E8BC0'> 2- Define Features and Target </span>

> X ==> Features 

> Y ==> Target 

```python
x = df[['SPX','USO','SLV','EUR/USD']]
y = df['GLD']
```

```python
x.head()
```

```python
y.head()
```

## <span style='color:#2E8BC0'> 3- Splitting Data </span>

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## <span style='color:#2E8BC0'> 4- Data Preprocessing </span>

```python
scaler = MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

## <span style='color:#2E8BC0'> 5- Building Model </span>

```python
model = Sequential()
```

```python
model.add(Dense(10,activation='relu',input_dim=4))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.compile(loss="mean_squared_error",optimizer='Adam')
```

```python
history = model.fit(x_train_scaled,y_train,epochs=50,validation_split=0.1)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
y_pred
```

## <span style='color:#2E8BC0'> 6- Model Evaluation </span>

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
